```python
import pandas as pd
```

```python
# Load weather data
df = pd.read_csv('rain.csv')

# Use only Central Park
df = df[df['STATION_NAME'] == 'NY CITY CENTRAL PARK NY US']

# Parse dates
df['weekofyear'] = pd.to_datetime(df['DATE'], format="%Y%m%d").dt.weekofyear
df['dayofweek'] = pd.to_datetime(df['DATE'], format="%Y%m%d").dt.dayofweek
df['weekend'] = df['dayofweek'] >= 5

# Did it rain?
df['precipitated'] = df['PRCP'] > 0

# Weekend days with Abby
abby = pd.read_csv('abby.csv')
abby['abby'] = True
df = df.merge(abby, on='DATE', how='left')

df
```

```python
# Average number of rainy days
all_days = df['precipitated'].mean()
all_days
```

```python
# Average number of rainy weekend days
weekend_days = df[df['weekend']]['precipitated'].mean()
weekend_days
```

```python
# Average number of rainy weekday days
week_days = df[~df['weekend']]['precipitated'].mean()
week_days
```

```python
# Average number of rainy weekend days when I saw Abby
abby_days = df[(df['weekend']) & (df['abby'])]['precipitated'].mean()
abby_days
```

```python
print(
    "It rains %s%% more often on weekend days when I see Abby than an average day in 2017" % int(((abby_days - all_days) / all_days) * 100)
)

print(
    "It rains %s%% more often on weekend days when I see Abby than an average weekend day in 2017" % int(((abby_days - weekend_days) / weekend_days) * 100)
)

print(
    "It rains %s%% less often on weekend days than an average weekday in 2017" % abs(int(((weekend_days - week_days) / week_days) * 100))
)
```
